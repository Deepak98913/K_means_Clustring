```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt


import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.datasets import make_blobs
X,y = make_blobs(n_samples=1000, centers=3, n_features=3, random_state=1)
```

```python
X
```

```python
y
```

```python
plt.scatter(X[:,0], X[:,1], c =y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y , test_size=0.30, random_state=1)
```

```python
X_train.shape, X_test.shape
```

```python
# How to make clusters
```

```python
from sklearn.cluster import KMeans
Kmeans = KMeans(n_clusters=3)
```

```python
Kmeans.fit(X_train, y_train)
```

```python
y_labels= Kmeans.predict(X_train)
```

```python
plt.scatter(X_train[:,0], X_train[:,1], c=y_labels)
```

```python
# How better the clustring model is

from sklearn.metrics import silhouette_score
silhouette_score(X_train, y_labels)
```

```python
Kmeans.labels_
```

```python
Kmeans.inertia_
```

```python
# to select K(centroid)>> elbow method

wcss = []
for k in range(1,10):
  kmeans = Kmeans(n_cluster = k)
  Kmeans.fit(X_train)
  wcss.append(kmeans.inertia_)
```

```python
wcss
```

```python
plt.plot(range(1,10), wcss)
plt.xticks(range(1,10))
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
```

```python
# based on the ablove plot the optimal no of k cluster is 3
```
